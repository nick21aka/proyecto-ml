# 01 — Business Understanding (CRISP–DM)

**Proyecto:** Predicción de riesgo de diabetes y cardiovascular  
**Objetivo:** construir un prototipo reproducible (Kedro) que entregue un **score de riesgo** y minimice **falsos negativos** para priorizar derivaciones clínicas.

## Stakeholders y uso
- Dirección médica, enfermería, analistas de datos, TI.  
- Decisión: derivar a control si **score ≥ umbral τ**.  
- Política de costos: **FN = 10**, **FP = 2**.

## Datasets
- `db_diabetes.csv` (target típico: `Outcome`)  
- `db_cardio.csv` (target típico: `cardio`)  
- `db_cardiabetes.csv` (target típico: `Diabetes`)  
> Si el target no existe, usar **fallback**: primera columna **binaria** hallada.

## Objetivos de ML (metas)
- **Recall ≥ 0.85** con **Precision ≥ 0.50** (en ≥ 2 de 3 datasets).  
- **AUC ROC ≥ 0.80**, **F1 ≥ 0.60** (deseable).  
- Selección de **umbral** por costo (FN≫FP).

## Parámetros sugeridos (`conf/base/parameters.yml`)
```yaml
targets:
  diabetes: Outcome
  cardio: cardio
  cardiabetes: Diabetes

split:
  test_size: 0.2
  random_state: 42

preprocessing:
  imputer_strategy_num: median
  imputer_strategy_cat: most_frequent
  scaler: standard
  categorical_encoding: onehot

costs:
  fn: 10
  fp: 2
